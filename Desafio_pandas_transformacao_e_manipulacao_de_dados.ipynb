```python
import pandas as pd
import numpy as np
```

# Projeto 1

```python
dados_clientes = pd.read_json('/content/dados_vendas_clientes.json')
dados_clientes.head()
```

```python
dados_clientes = pd.json_normalize(dados_clientes['dados_vendas'])
dados_clientes.head()
```

```python
dados_clientes = dados_clientes.explode(['Cliente','Valor da compra'])
dados_clientes.head()
```

```python
dados_clientes.reset_index(inplace=True, drop=True)
dados_clientes.head()
```

```python
dados_clientes['Valor da compra'] = dados_clientes['Valor da compra'].apply(lambda x: x.replace('R$ ','').replace(',','.'))
dados_clientes.head()
```

```python
dados_clientes['Valor da compra'] = dados_clientes['Valor da compra'].astype(np.float64)
```

```python
dados_clientes.info()
```

```python
dados_clientes.head()
```

```python
dados_clientes['Cliente'] = dados_clientes['Cliente'].str.lower().replace('[^a-zA-Z\ã]',' ',regex=True)
```

```python
dados_clientes.head()
```

```python
dados_clientes['Data de venda'] = pd.to_datetime(dados_clientes['Data de venda'], format='%d/%m/%Y')
```

```python
dados_clientes.info()
```

```python
dados_clientes
```

```python
dados_clientes['Data de venda'].dt.strftime('%d')
```

```python
subset = dados_clientes.groupby([dados_clientes['Cliente'], dados_clientes['Data de venda']])['Valor da compra'].sum()
```

```python
subset
```

```python
subset_clientes = subset.groupby('Cliente').sum(numeric_only=True)
```

```python
subset_clientes
```

# Projeto 2

```python
dados_locacao = pd.read_json('/content/dados_locacao_imoveis.json')
dados_locacao.head()
```

```python
dados_locacao = pd.json_normalize(dados_locacao['dados_locacao'])
```

```python
dados_locacao.head()
```

```python
colunas_locacao = list(dados_locacao.columns)
colunas_locacao
```

```python
dados_locacao = dados_locacao.explode(colunas_locacao[1:])
dados_locacao
```

```python
dados_locacao.reset_index(inplace=True, drop=True)
dados_locacao.head()
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].apply(lambda x: x.replace('$ ','').replace(',','.').replace(' reais',''))
dados_locacao.head()
```

```python
dados_locacao['valor_aluguel'] = dados_locacao['valor_aluguel'].astype(np.float64)
dados_locacao.info()
```

```python
dados_locacao['apartamento']
```

```python
dados_locacao['apartamento'] = dados_locacao['apartamento'].str.replace(' \(blocoAP\)','',regex=True)
```

```python
dados_locacao
```

```python
dados_locacao['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao['datas_combinadas_pagamento'], format='%d/%m/%Y')
```

```python
dados_locacao['datas_combinadas_pagamento']
```

```python
dados_locacao['datas_de_pagamento'] = pd.to_datetime(dados_locacao['datas_de_pagamento'], format='%d/%m/%Y')
```

```python
dados_locacao['datas_de_pagamento']
```

```python
pagamentos_atrasados = dados_locacao['datas_combinadas_pagamento'] < dados_locacao['datas_de_pagamento']
```

```python
subset_atrasos = dados_locacao.assign(pagamentos_atrasados=pagamentos_atrasados)
subset_atrasos
```
